```python
import cv2
import matplotlib.pyplot as plt
from sklearn import datasets
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
digits = datasets.load_digits()
plt.imshow(digits.images[0], cmap="gray")
plt.show()
```

```python
for i in range(15):
    plt.subplot(3, 5, i+1)
    plt.axis("off")
    plt.title(str(digits.target[i]))
    plt.imshow(digits.images[i], cmap="gray")

plt.show()
```

```python
x = digits.data
# x = x.image.reshape((-1, 64))

from sklearn.model_selection import train_test_split
from sklearn import datasets, svm, metrics
from sklearn.metrics import accuracy_score
from sklearn.linear_model import SGDClassifier

y = digits.target
x_train, x_test, y_train, y_test = train_test_split(x, y)

clf = svm.LinearSVC()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(accuracy_score(y_test, y_pred))

lin = SGDClassifier()
lin.fit(x_train, y_train)
y_pred = lin.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
# 학습한 모델을 저장
import joblib
joblib.dump(clf, 'digits.pkl')
joblib.dump(lin, 'digits-lin.pkl')
```

```python
def predict_digit(filename):
  # 학습한 데이터 읽어 들이기
    model = joblib.load("digits.pkl")
    my_img = cv2.imread(filename)
    my_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
    my_img = cv2.resize(my_img, (8, 8))
    my_img = 15 - my_img // 16 
    # 흑백 반전
    my_img = my_img.reshape((-1, 64))
    res = model.predict(my_img)
    return res[0]
```

```python
# 이미지 파일을 지정해서 실행하기
!curl https://raw.githubusercontent.com/StillWork/data/master/digit2.png\
    -o digit2.png
img = cv2.imread("digit2.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
n = predict_digit("digit2.png")
print("digit2.png = " + str(n))
plt.show()
```
